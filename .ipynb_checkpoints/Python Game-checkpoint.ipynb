```python
letters = { "a":[ "###", "# #", "###", "# #", "# #"], "b":[ "###", "# #", "###", "# #", "###"], "c":[ "###", "#", "#", "#", "###"], "d":[ "##", "# #", "# #", "# #", "##"], "e":[ "###", "#", "###", "#", "###"], "f":[ "###", "#", "###", "#", "#"], "g":[ "###", "# #", "###", "  #", "###"], "h":[ "# #", "# #", "###", "# #", "# #"], "i":[ "###", " #", " #", " #", "###"], "j":[ "###", " #", " #", " #", "##"], "k":[ "# #", "##", "#", "##", "# #"], "l":[ "#", "#", "#", "#", "###"], "m":[ "# #", "###", "###", "# #", "# #"], "n":[ "###", "# #", "# #", "# #", "# #"], "o":[ "###", "# #", "# #", "# #", "###"], "p":[ "###", "# #", "###", "#", "#"], "q":[ "###", "# #", "###", "  #", "  #"], "r":[ "###", "# #", "##", "# #", "# #"], "s":[ "###", "#", "###", "  #", "###"], "t":[ "###", " #", " #", " #", " #"], "u":[ "# #", "# #", "# #", "# #", "###"], "v":[ "# #", "# #", "# #", "# #", " #"], "w":[ "# #", "# #", "# #", "###", "###"], "x":[ "# #", " #", " #", " #", "# #"], "y":[ "# #", "# #", "###", "  #", "###"], "z":[ "###", "  #", " #", "#", "###"], " ":[ " "], "1":[ " #", "##", " #", " #", "###"], "2":[ "###", "  #", "###", "#", "###"], "3":[ "###", "  #", "###", "  #", "###"], "4":[ "#", "#", "# #", "###", "  #"], "5":[ "###", "#", "###", "  #", "###"], "6":[ "###", "#", "###", "# #", "###"], "7":[ "###", "  # ", " #", " #", "#"], "8":[ "###", "# #", "###", "# #", "###"], "9":[ "###", "# #", "###", "  #", "###"], "0":[ "###", "# #", "# #", "# #", "###"], "!":[ " # ", " # ", " # ", "   ", " # "], "?":[ "###", "  #", " ##", "   ", " # "], ".":[ "   ", "   ", "   ", "   ", " # "], "]":[ "   ", "   ", "   ", "  #", " # "], "/":[ "  #", "  #", " # ", "# ", "# "], ":":[ "   ", " # ", "   ", " # ", "   "], "@":[ "###", "# #", "## ", "#  ", "###"], "'":[ " # ", " # ", "   ", "   ", "   "], "#":[ " # ", "###", " # ", "###", " # "] }
# letters obtained from here: http://www.stuffaboutcode.com/2013/08/raspberry-pi-minecraft-twitter.html

def print_letters(text):
    bigletters = []
    for i in text:
        bigletters.append(letters.get(i.lower(),letters[' ']))
    output = ['']*5
    for i in range(5):
        for j in bigletters:
            temp = ' '
            try:
                temp = j[i]
            except:
                pass
            temp += ' '*(5-len(temp))
            temp = temp.replace(' ',' ')
            #temp = temp.replace('#','*')
            output[i] += temp
    return '\n'.join(output)
```

-> A function for some cool animations 

```python
# Function for implementing the loading animation 
def load_animation(str_in): 
  
    # String to be displayed when the application is loading 
    load_str = str_in
    ls_len = len(load_str) 
  
  
    # String for creating the rotating line 
    animation = "|/-\\"
    anicount = 0
      
    # used to keep the track of 
    # the duration of animation 
    counttime = 0        
      
    # pointer for travelling the loading string 
    i = 0                     
  
    while (counttime != 30): 
          
        # used to change the animation speed 
        # smaller the value, faster will be the animation 
        time.sleep(0.095)  
                              
        # converting the string to list 
        # as string is immutable 
        load_str_list = list(load_str)  
          
        # x->obtaining the ASCII code 
        x = ord(load_str_list[i]) 
          
        # y->for storing altered ASCII code 
        y = 0                             
  
        # if the character is "." or " ", keep it unaltered 
        # switch uppercase to lowercase and vice-versa  
        if x != 32 and x != 46:              
            if x>90: 
                y = x-32
            else: 
                y = x + 32
            load_str_list[i]= chr(y) 
          
        # for storing the resultant string 
        res =''              
        for j in range(ls_len): 
            res = res + load_str_list[j] 
              
        # displaying the resultant string 
        sys.stdout.write("\r"+res + animation[anicount]) 
        sys.stdout.flush() 
  
        # Assigning loading string 
        # to the resultant string 
        load_str = res 
  
          
        anicount = (anicount + 1)% 4
        i =(i + 1)% ls_len 
        counttime = counttime + 1
      
    # for windows OS 
    if os.name =="nt": 
        os.system("cls") 
          
    # for linux / Mac OS 
    else: 
        os.system("clear") 
```

A spawn function for when the game starts or the player loses life

```python
def spawn_player():
    seed()
    sequence = [3,4,5,6,16,22]
    return choice(sequence)
```

The start of the game 

```python
print("+-+-+-+-+-+-+-+-+-+-+-+-GAME RULES+-+-+-+-+-+-+-+-+-+-+-+-\n")
print("1) Your objective is to make it from your spawned area to the end of the map")
time.sleep(0.2)
print("2) The game will not display the map to you, you have to graph it out yourself, best way is trial & error")
time.sleep(0.2)
print("3) You have 3 lives to win the game")
time.sleep(0.2)
print("4) Every time you die, you will be spawned to a random spawnable area")
time.sleep(0.2)
print("5) All the best")
time.sleep(0.2)


print("Press any Enter to start the game")
input()

name = input("Enter Player Name : ")
move_count = 0;

load_animation("Starting the game ... ")
print("\n")
print(f"All the best {name} !!")
time1 = datetime.now()
lives=3
loc = spawn_player()

while lives>0:
        
    print(f"\nYou find yourself in {re.split('[0-9]',locations[loc])[0]}")
    
    if loc in death_locations.keys():
        time.sleep(1)
        print("You experience am eerie feeling!!")
        load_animation("Ahhh!!")
        #time.sleep(1)
        print("\n")
        print(f"\nOh no!!! \n\n{death_locations[loc]}\n")
        load_animation("Healing player")
        print("\n")
        lives=lives-1
        print(f"\nYou have {lives} lives left\n")
        loc = spawn_player()
    
    elif loc==12:
        time.sleep(1)
        print("\nYou hear a distant thudding noise !! ")
        load_animation("...")
        print("\nOh no!!! You see a large army coming closer!\n What could their intentions be?\n")
        load_animation("...")
        print("\nTheir big and burly army general calls you king and bows down!!! Is this a ploy to earn your trust???\n")
        load_animation("...")
        print("\n")
        print(print_letters(" YOU WIN !!! "))
        print("\n")
        print("\n")
        break
    
    else:
        objects_in_surrounding = {}
        count=0
        
        #Iterating
        for i in gameMap[loc]:
            if i>0:
                objects_in_surrounding[i] = count
            count+=1
                
        for i in range(1,9):
            if i in objects_in_surrounding.keys():
                print(f"\nIn your {directions[i]}({i}) lies {re.split('[0-9]',locations[objects_in_surrounding[i]])[0]}")
        
        user_input=""
        
        while True:
            try:
                user_input = int(input("\nWhich direction do you want to go?   "))
            
            except ValueError: 
                print("The entered value is not a number! please try again ")
                
            else:
                
                if user_input in objects_in_surrounding.keys():
                    time.sleep(0.5)
                   # os.system('cls')
                    clear_output()
                    print('\n-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+')
                    print("\nYou move towards your "+ directions[user_input])
                    loc=objects_in_surrounding[user_input]
                    time.sleep(0.5)
                    move_count = move_count + 1
                    break
                
                else: 
                    print("\nYour path is blocked in that direction!")
                    #os.system('cls')
                    #clear_output()
                    time.sleep(1)

print(print_letters("GAME OVER\n\n"))
time2 = datetime.now()
time_final = time2-time1
minutes = math.floor(int(time_final.seconds)/60)

print("\n\n\n+-+-+-+-+-+-+-+-+-+-+-+-GAME STATS+-+-+-+-+-+-+-+-+-+-+-+-\n")
print(f"Player Name : {name}")
print(f"Number of lives left : {lives}")
print(f"Time played : {(minutes)}:{time_final.seconds}")
print(f"No of moves : {move_count}")
                      
print("\n\n\n+-+-+-+-+-+-+-+-+-+-+-+-LEADERBOARD+-+-+-+-+-+-+-+-+-+-+-+-\n")
scores = open('scores.txt','a+')
scores.write(f"{name}                  {minutes}:{time_final.seconds}                               {move_count}                              {lives}\n")
scores.seek(0) 
data = scores.read()
print(data)
scores.close()
                      
      
```
